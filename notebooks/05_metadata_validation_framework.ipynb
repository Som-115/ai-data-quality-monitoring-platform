# Metadata Driven Validation Framework

## Version 2 - Phase 1

In this notebook, we are building a metadata-driven validation engine for dynamic rule execution.

Objectives:
- create reusable validation framework
- store validation rules in metadata tables
- dynamically apply validation logic
- reduce hardcoded validation implementations
- improve scalability and maintainability

This framework enables enterprise-grade configurable data quality validations.

#### Step 1: Import Required Libraries

```python
from pyspark.sql.functions import *
from pyspark.sql.types import *
```

#### Step 2: Create Validation Metadata Rules

We are creating configurable validation rules for dynamic validation execution.

```python
validation_rules = [

    ("taxi_trips", "fare_amount", "greater_than", "0", "HIGH"),

    ("taxi_trips", "trip_distance", "greater_than", "0", "HIGH"),

    ("taxi_trips", "passenger_count", "between", "1,6", "MEDIUM"),

    ("taxi_trips", "payment_type", "not_null", "NA", "HIGH"),

    ("taxi_trips", "tpep_pickup_datetime", "not_null", "NA", "HIGH")

]

```

#### Step 3: Create Validation Rules DataFrame

```python
schema = StructType([
    StructField("table_name", StringType(), True),
    StructField("column_name", StringType(), True),
    StructField("validation_type", StringType(), True),
    StructField("rule_value", StringType(), True),
    StructField("severity", StringType(), True)
])
```

```python
validation_rule_df = spark.createDataFrame(validation_rules, schema)
```

```python
validation_rule_df.show()
```

#### Step 4: Store Validation Rules Metadata Table

We are storing configurable validation rules as Delta tables for dynamic validation processing.

```python
validation_rule_df.write\
    .format("delta")\
        .mode("overwrite")\
            .saveAsTable("taxi_quality_platform.config.validation_rules")
```

#### Step 5: Verify Metadata Rules Table

```python
rules_df = spark.read.table("taxi_quality_platform.config.validation_rules")
```
