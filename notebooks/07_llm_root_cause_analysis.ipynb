#### Step 2: Import Required Libraries

```python
import google.generativeai as genai

from pyspark.sql.functions import *
from pyspark.sql.types import *
```

#### Step 3: Configure Gemini API

```python
GEMINI_API_KEY = GEMINI_API_KEY = input("Enter Gemini API Key: ")
```

```python
genai.configure(api_key=GEMINI_API_KEY)
```

#### Step 4: Initialize Gemini Model

```python
for model in genai.list_models():
    print(model.name)
```

```python
model = genai.GenerativeModel(
    "models/gemini-2.5-flash"
)
```

#### Step 5: Load Dynamic Validation Summary

```python
validation_summary_df = spark.table("taxi_quality_platform.gold.dynamic_validation_summary")
```

#### Step 6: Filter Failed Validations

```python
failed_validation_df = validation_summary_df.filter(col("failed_count") > 0)
failed_validation_df.show()
failed_validation_df.count()
```

#### Step 7: Convert Failed Validations into Python Objects

```python
failed_validations = failed_validation_df.collect()
```

```python
print(failed_validations)
```

#### Step 8: Generate AI Root Cause Analysis Using Gemini

```python
ai_incidents = []
```

```python
for validation in failed_validations:

    validation_column = validation["column"]

    failed_count = validation["failed_count"]


    prompt = f"""
    
    You are an enterprise data reliability engineer.

    Analyze the following validation failure:

    Validation Column: {validation_column}

    Failed Record Count: {failed_count}

    Generate:
    1. Probable Root Cause
    2. Business Impact
    3. Severity Level
    4. Recommended Resolution

    Keep response concise and professional.
    
    """


    response = model.generate_content(prompt)


    ai_response = response.text


    ai_incidents.append(

        (
            validation_column,
            failed_count,
            ai_response
        )

    )
```

#### Step 9: Create AI Incident Analysis DataFrame

```python
schema = StructType([
    StructField("column", StringType()),
    StructField("failed_count", LongType()),
    StructField("ai_response", StringType())
])
```

```python
ai_incidents_df = spark.createDataFrame(ai_incidents, schema)
display(ai_incidents_df)
```

#### Step 10: Store AI Incident Analysis Table

```python
ai_incidents_df.write \
    .format("delta") \
    .mode("overwrite") \
    .saveAsTable(
        "taxi_quality_platform.gold.ai_incident_analysis"
    )
```
